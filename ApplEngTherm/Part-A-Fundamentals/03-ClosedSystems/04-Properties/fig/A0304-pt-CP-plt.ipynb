# $c_P(T)$ Plotting Notebook

## Imports

```julia
using CSV
```

```julia
using DataFrames
```

```julia
using Plots
```

## Load Data

```julia
df = DataFrame(CSV.File("A0304-pt-CP.csv"));
first(df, 5)
```

```julia
RU = 8.31447
```

## Plotting

```julia
theme(:gruvbox_dark)
```

```julia
default(
    titlefont = "times",
    legendfont = "times",
    guidefont = "times",
    tickfont = "times",
)
```

```julia
p1 = plot(
    df.T,
    df.He ./ RU,
    size = (480, 270), # = 30 .* (16, 9)
    label = "He",
    lw = 1.5,
    xlabel = "T/K",
    ylabel = "c̄P/R̄",
    gridlinewidth = 0.6,
    xminorgrid = true,
    yminorgrid = true,
    leg = true,
    xticks = [(0:500:3500)...],
    yticks = 0.0:1.0:8.0,
    xlims = (100.0, 3900.0),
    ylims = (  2.0,    8.0),
    xrotation = 0.0,
)
plot!(p1, df.T, df.H₂  ./ RU, label="H₂", lw = 1.5)
plot!(p1, df.T, df.ar  ./ RU, label="ar", lw = 1.5)
plot!(p1, df.T, df.O₂  ./ RU, label="O₂", lw = 1.5)
plot!(p1, df.T, df.H₂O ./ RU, label="H₂O", lw = 1.5)
plot!(p1, df.T, df.CO₂ ./ RU, label="CO₂", lw = 1.5)
```

```julia
savefig("A0304-pt-CP.svg")
```
